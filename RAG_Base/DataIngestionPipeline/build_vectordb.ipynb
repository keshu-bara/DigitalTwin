```python
import json
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
#load embedding model

embedding_model = HuggingFaceEmbeddings(
    model_name = "sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# load personality memory dataset
with open("../data/rag_memory/processeddata.json","r",encoding = "utf-8") as f:
    data = json.load(f)

documents = []

for item in data:
    doc = Document(
        page_content = item["page_content"],
        metadata=item["metadata"]
    )
    documents.append(doc)


print("Document loaded:", len(documents))


```

```python
# create vector data base
vectorstore = Chroma.from_documents(
    documents,
    embedding_model,
    persist_directory = "./personality_vector_db"
)
```

```python
vectorstore.persist()

print("Vectore database created successfully.")
```
