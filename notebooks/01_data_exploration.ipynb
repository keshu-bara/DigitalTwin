```python
import re
import json
chats = '../data/NikhilMishraCE.txt'
```

```python
with open(chats,"r",encoding = 'utf-8') as f:
    lines = f.readlines()

```

```python
lines[0].split("-",1)
```

```python
messages = []
current_message = None

mss_pattern = r"^\d{2}/\d{2}/\d{4},\s\d{1,2}:\d{2}.*?-" #detecting starting pattern

for line in lines:
    line = line.strip()

    if re.match(mss_pattern,line):

        if current_message: #saving previous message before starting new one
            messages.append(current_message)

        # Split timestamp and rest
        parts = line.split("-",1)

        timestamp = parts[0]

        # Split speaker and text
        speaker_text = parts[1].split(": ",1)

        if len(speaker_text) < 2:
            continue # skip system messages

        speaker = speaker_text[0]

        text = speaker_text[1]

        # skip media messages

        if "<Media omitted>" in text:
            continue

        current_message = {
            "timestamp": timestamp,
            "speaker": speaker,
            "text": text
        }

    else:
        # continuation of previous message
        if current_message:
            current_message["text"] += " " + line


# Append last message

if current_message:
    messages.append(current_message)


        

            
```

```python
messages
```

```python
with open("../data/processed/messages.json", "w", encoding="utf-8") as f:
    json.dump(messages, f, ensure_ascii = False, indent = 4)
```
