```python
with open("../data/processed/messages.json","r",encoding="utf-8") as f:
    messages = json.load(f) #loading json formate as python dic formate

len(messages)
```

```python
merged_messages = []

current_block = messages[0]

for i in range(1, len(messages)):

    msg = messages[i]

    # if same speaker as previous -> merge it
    if msg['speaker'] == current_block['speaker']:
        current_block['text'] += " " + msg["text"]

    else:
        # Different speaker -> save the previous block
        merged_messages.append(current_block)
        current_block = msg

# Append last block
merged_messages.append(current_block)

print(len(merged_messages),len(messages))
print(merged_messages)
```
